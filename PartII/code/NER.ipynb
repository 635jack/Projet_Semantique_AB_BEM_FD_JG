```python
data = pd.read_csv("data_with_gliner.csv")
data.head()
```

```python
import ast
def clean_numpy_array_string(text):
    if not isinstance(text, str): return text
    clean_text = text.replace('[', '').replace(']', '').strip()
    if not clean_text: return []
    return [int(x) for x in clean_text.split()]

data['ner_tags'] = data['ner_tags'].apply(clean_numpy_array_string)
data['gliner_entities'] = data['gliner_entities'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
data['words'] = data['text'].apply(lambda x: x.split(' ') if isinstance(x, str) else [])
```

# Gliner Evaluation

```python
import pandas as pd
import ast

cols_to_fix = ['words', 'ner_tags', 'gliner_entities']

for col in cols_to_fix:
    data[col] = data[col].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
true_label_map = {
    0: 'O',
    1: 'LOC',
    2: 'PER',
    3: 'MISC',
    4: 'ORG'
}

gliner_label_map = {
    'person': 'PER',
    'location': 'LOC',
    'organization': 'ORG',
    'miscellaneous': 'MISC'

}
```

```python
def align_gliner_to_tokens(row):
    words = row['words']
    gliner_ents = row['gliner_entities']

    token_pred_tags = ['O'] * len(words)

    if not gliner_ents:
        return token_pred_tags

    current_char_idx = 0
    token_spans = []
    for word in words:
        start = current_char_idx
        end = start + len(word)
        token_spans.append((start, end))
        current_char_idx = end + 1


    for ent in gliner_ents:
        entity_start = ent['start']
        entity_end = ent['end']
        label_long = ent['label']

        label_short = gliner_label_map.get(label_long, 'O')

        for idx, (tok_start, tok_end) in enumerate(token_spans):
            tok_mid = (tok_start + tok_end) / 2
            if entity_start <= tok_mid < entity_end:
                token_pred_tags[idx] = label_short

    return token_pred_tags
```

```python
data['aligned_preds'] = data.apply(align_gliner_to_tokens, axis=1)

data['aligned_true'] = data['ner_tags'].apply(lambda tags: [true_label_map.get(t, 'O') for t in tags])
```

```python
from sklearn.metrics import classification_report

y_true = [tag for row in data['aligned_true'] for tag in row]
y_pred = [tag for row in data['aligned_preds'] for tag in row]

labels_to_eval = ['PER', 'LOC', 'ORG', 'MISC']

print("Evaluation Report")
print(classification_report(y_true, y_pred, labels=labels_to_eval))
```

# Second Model: Spacy

```python
#!pip install spacy
#!python -m spacy download en_core_web_sm
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")


spacy_to_wikiner_map = {
    'PERSON': 'PER',
    'ORG': 'ORG',
    'GPE': 'LOC',
    'LOC': 'LOC',
    'NORP': 'MISC',
    'FAC': 'LOC',

}

def extract_spacy_entities(text):
    if not isinstance(text, str):
        return []

    doc = nlp(text)
    entities = []

    for ent in doc.ents:
        if ent.label_ in spacy_to_wikiner_map:
            entities.append({
                'text': ent.text,
                'start': ent.start_char,
                'end': ent.end_char,
                'label': spacy_to_wikiner_map[ent.label_]
            })

    return entities


print("Extraction SpaCy en cours...")
tqdm.pandas()

data['spacy_entities'] = data['cleaned_text'].progress_apply(extract_spacy_entities)
```

```python
def align_spacy_to_tokens(row):
    words = row['words']
    spacy_ents = row['spacy_entities']

    token_pred_tags = ['O'] * len(words)

    if not spacy_ents:
        return token_pred_tags

    current_char_idx = 0
    token_spans = []
    for word in words:
        start = current_char_idx
        end = start + len(word)
        token_spans.append((start, end))
        current_char_idx = end + 1

    for ent in spacy_ents:
        entity_start = ent['start']
        entity_end = ent['end']
        label = ent['label']

        for idx, (tok_start, tok_end) in enumerate(token_spans):
            tok_mid = (tok_start + tok_end) / 2
            if entity_start <= tok_mid < entity_end:
                token_pred_tags[idx] = label

    return token_pred_tags

data['spacy_aligned_preds'] = data.apply(align_spacy_to_tokens, axis=1)
```

```python
from sklearn.metrics import classification_report

y_true = [tag for row in data['aligned_true'] for tag in row]
y_pred_spacy = [tag for row in data['spacy_aligned_preds'] for tag in row]

labels_to_eval = ['PER', 'LOC', 'ORG', 'MISC']

print(" Evaluation Spacy")
print(classification_report(y_true, y_pred_spacy, labels=labels_to_eval, digits=4))
```
